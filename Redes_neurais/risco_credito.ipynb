```python
from sklearn.neural_network import MLPClassifier
```

```python
import pickle
with open('../PKLs/credit.pkl', 'rb') as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape
```

## Explicando os parâmetros

max_iter - o próprio nome entrega a funcionalidade
verbose - responsável pelos prints abaixo
tol - parâmetro de tolerância explicitado nos prints, para entender melhor, observe a última linha dele

```python
rede_neural_credit = MLPClassifier(max_iter=1500, verbose=True, tol=0.000010)
rede_neural_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = rede_neural_credit.predict(X_credit_teste)
previsoes
```

```python
y_credit_teste
```

## Porcentagem de acurácia

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_credit_teste, previsoes)
```

## Classification report ( por muito o melhor até agora! )

```python
print(classification_report(y_credit_teste, previsoes))
```

## UM ERRO EM CADA CASO APENAS LLLLKKKKKKKKKKKKKKK

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(rede_neural_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste, y_credit_teste)
```
